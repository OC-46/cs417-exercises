```python
from math import gcd

class Fraction:
    def __init__(self, top, bottom):
        self.num = top
        self.den = bottom

    def __str__(self):
        return f"{self.num}/{self.den}"

    def _reduce(self):
        """Reduce fraction to lowest terms using GCD"""
        common_divisor = gcd(abs(self.num), abs(self.den))
        self.num //= common_divisor
        self.den //= common_divisor
        return self

    def __add__(self, other):
        """Add two fractions: a/b + c/d = (ad + bc)/(bd)"""
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den
        result = Fraction(new_num, new_den)
        return result._reduce()

    def __sub__(self, other):
        """Subtract two fractions: a/b - c/d = (ad - bc)/(bd)"""
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den
        result = Fraction(new_num, new_den)
        return result._reduce()

    def __mul__(self, other):
        """Multiply two fractions: (a/b) * (c/d) = (ac)/(bd)"""
        new_num = self.num * other.num
        new_den = self.den * other.den
        result = Fraction(new_num, new_den)
        return result._reduce()

    def __truediv__(self, other):
        """Divide two fractions: (a/b) / (c/d) = (a/b) * (d/c) = (ad)/(bc)"""
        new_num = self.num * other.den
        new_den = self.den * other.num
        result = Fraction(new_num, new_den)
        return result._reduce()


    #test cases

    #test case 1
    f1 = Fraction(3,5)
    f2 = Fraction(1,3)
    result_a = f1 + f2
    print(f"{f1} + {f2} = {result_a}")

    #test case 2
    result_m = f1 * f2
    print(f"{f1} * {f2} = {result_m}")

    #test case 3
    result_d = f1 / f2
    print(f"{f1} / {f2} = {result_d}")

    #test case 4
    result_s = f1 - f2
    print(f"{f1} - {f2} = {result_s}")
```
